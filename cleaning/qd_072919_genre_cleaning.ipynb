## Import genre list

Import the genre list provided with the data. In our tables, the genre names have bo spaces (but the genre list does). Take the genre lists and compute what the genres would be without spaces. We will later store these in our actual dataset. 

```python
genre_df = pd.read_csv("../data/raw/genre_dataset.txt")
genre_df["Condensed_name"] = genre_df.Genre.apply(lambda x: x.replace(" ", ""))
genre_df.set_index("Condensed_name", inplace=True)
genre_df.head()
```

Here, we take our genre list and match the expanded name to the condensed names in our main dataset.

```python
data['Expanded_genre'] = np.NaN

def match_expanded_genre(x):
    try:
        return genre_df.loc[x].Genre
    except:
        return np.NaN

data.Expanded_genre = data.Genre.apply(match_expanded_genre)
```

reorder our columns to place genres next to each other. 

```python
data = data[[
    'Name',
    'Danceability',
    'Energy',
    'Key',
    'Loudness',
    'Mode',
    'Speechness',
    'Acousticness',
    'Instrumentalness',
    'Liveness',
    'Valence',
    'Tempo',
    'Type',
    'ID',
    'Uri',
    'Ref_Track',
    'URL_features',
    'Duration_ms',
    'time_signature',
    'Genre',
    'Expanded_genre',
    'artist_names',
    'artist_ids',
    'release_date',
    'release_date_precision',
]]
data.head()
```

## Add super genre column

```python
data[data.Expanded_genre.str.contains("grunge")].Expanded_genre.value_counts()
```

```python
super_genres = [
    "pop",
    "rock",
    "r&b",
    "country",
    "soul",
    "latin",
    "disco",
    "metal",
    "dance",
    "jazz",
    "synth",
    "hiphop",
    "folk",
    "swing",
    "blues",
    "reggae",
    "grunge",
    "classical",
    "indie",
    "elect",
    "techno",
    "step",
    "punk",
    "worship",
    "guitar",
    "house",
    "edm" 
]
len(super_genres)
```

```python
total = 0
for g in super_genres:
    genre_count = data[data.Expanded_genre.str.contains(g)].Expanded_genre.value_counts().sum()
    total += genre_count
    print(g, genre_count)
    
print(total)
```

```python
list(data.Expanded_genre.value_counts().index)
```

```python
def assign_super_genre(x):
    result = "general"
    for sg in super_genres:
        if sg in x:
            result = sg
            
    return result

    
data["Super_genre"] = data.Genre.apply(assign_super_genre)
```

```python
data[data.Super_genre != "general"].Super_genre.value_counts().sort_values()
```

```python
data.to_csv("../data/interum/top_genres.csv", index=False)
```
